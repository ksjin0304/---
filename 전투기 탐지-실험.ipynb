```python
import ultralytics
from ultralytics import YOLO
```

```python
ultralytics.checks()
```

```python
# Create a YOLOv8 model
model = YOLO('yolov8s.pt')
```

```python
#하이퍼 파라미터 설정
#하이퍼 파라미터 설정
epochs = 100
patience = 10
batch = 64
imgsz = 640
Optimizer = 'SGD'
dropout= 0.2
```

```python
model.train(data= '/content/drive/MyDrive/fighter_detection/mad.yaml', epochs= epochs, batch= batch, imgsz= imgsz, optimizer= Optimizer, patience= patience, dropout=dropout)
```

```python
results = model.predict(source = '/content/drive/MyDrive/fighter_detection/test/images', save=True)
```

```python
model = YOLO('/content/drive/MyDrive/전투기탐지결과2/train/weights/best.pt')
```

```python
#validation set에 대한 성능평가
val_metrics = model.val(split = 'val')
print('map:', val_metrics.box.map)
print('map50:', val_metrics.box.map50)
print('map75:', val_metrics.box.map75)
print('maps:', val_metrics.box.maps)
```

```python
#test set에 대한 성능평가
test_metrics = model.val(split = 'test')
print('map:', test_metrics.box.map)
print('map50:', test_metrics.box.map50)
print('map75:', test_metrics.box.map75)
print('maps:', test_metrics.box.maps)
```
